```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
plt.ion()
import time
```

```python
networks = { 'Dense': 'dense',
             'Conv1': 'conv1',
             'Conv2': 'conv2',
             'Inception': 'inception',
             'ResNet': 'resnet'}

startPath = '/s/red/b/nobackup/data/bsattelb/linearRegions/'
endPath = 'MNIST{}Region{}.npy'

for key in networks:
    networks[key] = startPath + networks[key] + endPath
keyLabels = list(networks.keys())
```

```python
def calcUniqueRegions(regions):
    regionsSet = []
    for i in range(regions.shape[0]):
        isGood = True
        for elem in regionsSet:
            if np.linalg.norm(regions[i, :] - elem) < 1e-3:
                isGood = False
                break
        if isGood:
            regionsSet.append(regions[i, :])
            
        if i % 100 == 0:
            print('\t{}, {}'.format(i, len(regionsSet)))
            
    return len(regionsSet)
```

```python
for key in keyLabels:
    for i in range(1):
        print(key, i)
        denseRegions = np.load(networks[key].format('Test', i))
        print(calcUniqueRegions(denseRegions))
```
